```python
#import packages
import pandas as pd
import plotly.express as px
import numpy as np
from statsmodels.tsa.stattools import adfuller #for testing stationarity
```

```python
#specify path for reading dataset
path = "./data/AirPassengers.csv"
```

```python
#read input data from above path
data = pd.read_csv(path)
data.head(5)
```

```python
from re import template
#define function to plot time series
def ts_plot(title, data, x, y, x_label, y_label):
  #generate basic chart
  fig = px.line(data, x= data[x], y = data[y], labels= {x:x_label, y:y_label})
  #format chart
  fig.update_layout(
      template = "simple_white",
      font = dict(size=18),
      title_text = title,
      width = 650,
      height = 400,
      title_x = 0.5
  )
  #display chart
  fig.show()
```

```python
#plot count of passengers by date using above plot function
ts_plot(title = "Airline Passengers Count by Date", data = data, x = "Month", y = "#Passengers",
        x_label= "Month", y_label = "Passengers Count" )
```

The plot is clearly not stationary since it has an upward trend and strong seasonality pattern
